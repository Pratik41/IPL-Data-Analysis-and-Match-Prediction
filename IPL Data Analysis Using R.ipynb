```r
df = subset(match, select = -c(id,city,date,result,dl_applied,win_by_runs,win_by_wickets,player_of_match,venue,umpire1,umpire2))
```

# Splitting the data into train and test

```r
set.seed(100)
train <- sample(nrow(df), 0.7*nrow(df), replace = FALSE)
TrainSet <- df[train,]
TestSet <- df[-train,]
```

```r
TrainSet$season <- as.factor(TrainSet$season)
TrainSet$winner <- as.factor(TrainSet$winner)
TrainSet$team1 <- as.factor(TrainSet$team1)
TrainSet$team2 <- as.factor(TrainSet$team2)
TrainSet$toss_winner <- as.factor(TrainSet$toss_winner)
TrainSet$toss_decision <- as.factor(TrainSet$toss_decision)

TestSet$season <- as.factor(TestSet$season)
TestSet$winner <- as.factor(TestSet$winner)
TestSet$team1 <- as.factor(TestSet$team1)
TestSet$team2 <- as.factor(TestSet$team2)
TestSet$toss_winner <- as.factor(TestSet$toss_winner)
TestSet$toss_decision <- as.factor(TestSet$toss_decision)
```

```r
levels(TestSet$season) <- levels(TrainSet$season)
levels(TestSet$winner) <- levels(TrainSet$winner)
levels(TestSet$team1) <- levels(TrainSet$team1)
levels(TestSet$team2) <- levels(TrainSet$team2)
levels(TestSet$toss_winner) <- levels(TrainSet$toss_winner)
levels(TestSet$toss_decision) <- levels(TrainSet$toss_decision)
```

# Fitting the Random Forest model

```r
model1 <- randomForest(winner~. , data = TrainSet, importance = TRUE)
```

```r
y_pred = predict(model1, newdata = TestSet[-6])
```
